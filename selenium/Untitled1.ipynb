```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from PIL import Image, ImageChops
import os
import time

# Caminho onde o Chrome est√° logado
# (ajuste se necess√°rio ‚Äî opcional)
#chrome_user_data = os.path.expanduser("~/.config/google-chrome")

# Configura√ß√µes do Chrome
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument("--disable-infobars")
chrome_options.add_argument("--disable-extensions")
#chrome_options.add_argument(f"--user-data-dir={chrome_user_data}")  # j√° logado

# Iniciar WebDriver
#driver.set_window_size(1920, 1080)  # Ou at√© maior
driver = webdriver.Chrome(options=chrome_options)
```

```python
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
from io import BytesIO
from reportlab.lib.utils import ImageReader
```

```python
os.makedirs("slides", exist_ok=True)
actions = ActionChains(driver)

output_pdf = "slides/slides_apresentacion.pdf"

c = None

contador = 0

while True:
    filename = f"slides/slide_{contador+1:02}.png"
    driver.save_screenshot(filename)
    print(f"‚úÖ Capturado: {filename}")

    if contador > 0:
        anterior = f"slides/slide_{contador:02}.png"
        if imagens_sao_iguais(anterior, filename):
            print("‚ö†Ô∏è Slide repetido detectado. Fim da apresenta√ß√£o.")
            os.remove(filename)
            break

    # üñºÔ∏è Redimensionar imagem para preencher p√°gina A4
    img = Image.open(filename).convert("RGB")
    
    img_width, img_height = img.size
    
    # Criar o canvas na primeira itera√ß√£o
    if c is None:
        c = canvas.Canvas(output_pdf, pagesize=(img_width, img_height))
    
    # Adicionar imagem e links
    c.drawImage(ImageReader(img), 0, 0, width=img_width, height=img_height)

    
    # üîó Detectar e adicionar links reais
    links = driver.find_elements(By.XPATH, "//a[@href]")
    for link in links:
        try:
            loc = link.location
            size = link.size
            win_width = driver.execute_script("return window.innerWidth")
            win_height = driver.execute_script("return window.innerHeight")

            x1 = (loc['x'] / win_width) * page_width
            y1 = page_height - (loc['y'] / win_height) * page_height
            x2 = x1 + (size['width'] / win_width) * page_width
            y2 = y1 - (size['height'] / win_height) * page_height

            c.linkURL(link.get_attribute("href"), (x1, y2, x2, y1), relative=0)
        except Exception as e:
            print("‚ö†Ô∏è Erro ao inserir link:", e)

    c.showPage()
    contador += 1
    actions.send_keys(Keys.ARROW_RIGHT).perform()
    time.sleep(1.5)

c.save()
driver.quit()
print(f"üìÑ PDF salvo com {contador} slides: '{output_pdf}'")
```
