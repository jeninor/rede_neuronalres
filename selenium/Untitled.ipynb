```python
import os
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from PIL import Image, ImageChops
```

```python
# Criar pasta para salvar os slides
os.makedirs("slides", exist_ok=True)




# Esperar o primeiro slide carregar
time.sleep(5)

actions = ActionChains(driver)

def imagens_sao_iguais(img1_path, img2_path):
    img1 = Image.open(img1_path)
    img2 = Image.open(img2_path)
    diff = ImageChops.difference(img1, img2)
    return not diff.getbbox()  # True se são iguais

i = 0
repetido = False

while not repetido:
    current_filename = f"slides/slide_{i+1:02}.png"
    driver.save_screenshot(current_filename)
    print(f"✅ Capturado: {current_filename}")

    if i > 0:
        anterior_filename = f"slides/slide_{i:02}.png"
        if imagens_sao_iguais(anterior_filename, current_filename):
            print("⚠️ Slide repetido detectado. Fim da apresentação.")
            os.remove(current_filename)  # Remove slide duplicado
            break

    actions.send_keys(Keys.ARROW_RIGHT).perform()
    time.sleep(1)
    i += 1

driver.quit()
print("✅ Finalizado. Slides salvos na pasta 'slides/'.")
```
